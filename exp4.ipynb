```python
#set theory union:OR and intersection:AND cant be used in lists and tpl
A={1,2,3,4}
B={3,4,5,6}
#union
print("union:",A|B)
print("intersection:",A&B)
print("intersection m2",A.intersection(B))#m2
#qu se understandwhat to use set tpl or list
print("difference:",A-B)
print("difference:",B-A)
#XOR
print("dissimilar:",A^B)
```

```python
'''Problem Statement3: One set contains students in the Cricket club and another in the football club.
Find students who are:
1. In both clubs
2.In only one club
Concept used:intersection,symmertric difference'''
cricket={"ram","subh","rishi","harsh","vabhay","aryan"}
football={"prateek","arjun","kunal","riya","aryan","ram","subh"}
print("Students in both clubs:",cricket&football)
print("Students in only one club:",cricket^football)
```

```python
'''Problem statement 4: Absent students
  Given a set of all students and set of present students,find the absent students.'''
students={"ram","sham","raywant","priyal","riya","shravya","rahul","priya"}
present={"ram","sham","raywant","riya","shravya","rahul"}
print("absent students:",students-present)
```
